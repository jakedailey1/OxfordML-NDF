```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import inspect
import time
from IPython.display import clear_output
```

```python
from tensorflow.examples.tutorials.mnist import input_data
from tensorflow.examples.tutorials.mnist import mnist
```

```python
np.set_printoptions(threshold=np.nan)
np.set_printoptions(suppress=True)
```

#### Input feeding method borrowed from: https://github.com/tensorflow/tensorflow/blob/master/tensorflow/examples/tutorials/mnist/fully_connected_feed.py

```python
graph = tf.Graph()
sess = tf.InteractiveSession(graph=graph)
```

```python
def fill_feed_dict(data_set, image_pl, label_pl,batch_size):
    # Create the feed_dict for the placeholders filled with the next
    # `batch size` examples.
    images_feed, labels_feed = data_set.next_batch(batch_size)
    feed_dict = {
      image_pl: images_feed,
      label_pl: labels_feed
    }
    return feed_dict
```

```python
data_sets = input_data.read_data_sets('MNIST_data',one_hot = True)
```

#### Initializing placeholders for the data

```python
x = tf.placeholder(tf.float32, shape=[None, 28*28])
y_ = tf.placeholder(tf.float32, shape = [None, 10])
```

#### Initializing weights  for the linear layer

```python
W1 = tf.Variable(tf.random_normal([784,10]).eval(), name='weights', trainable=True)
b1 = tf.Variable(tf.random_normal([10]).eval(), name = 'bias', trainable = True)
```

```python
sess.run(tf.global_variables_initializer())
```

#### Defining a batch size here so we can see what the code is doing along the way

```python
batch_size = 10
```

```python
temp_dict = fill_feed_dict(data_sets.train, x, y_, batch_size)
```

### Defining softmax and a loss function

#### Starting with a linear layer/transformation of pixels and parameters (+ a bias term)

```python
model_y = tf.matmul(x,W1) + b1
```

```python
sess.run(model_y, feed_dict=temp_dict)
```

#### Now feeding the linear layer into the softmax layer

```python
smax_num = tf.transpose(tf.exp(model_y-tf.reduce_max(model_y)))
smax_den = tf.reduce_sum(tf.exp(model_y-tf.reduce_max(model_y)),-1)
```

```python
softmax_y = tf.transpose(tf.divide(smax_num, smax_den))
sess.run(softmax_y, feed_dict=temp_dict)
```

#### Define Negative Log-Likelihood criterion as the element-wise product of one-hot labels and the log(softmax(x, weights)); this gives - log(p(label | image, weights)), a value we can minimize to achieve a MAP estimate of the model weights given a uniform prior.

```python
NLLCriterion = -tf.reduce_mean(tf.reduce_sum(tf.multiply(y_,tf.log(softmax_y+1e-10)),axis=1))
sess.run(NLLCriterion, feed_dict=temp_dict)
```

#### Finding the gradient of the loss w.r.t. our parameters

```python
dl_dW = tf.gradients(NLLCriterion, [W1])[0]
```

```python
sess.run(dl_dW,feed_dict = temp_dict)
```

#### Finding the gradient of the loss w.r.t. the data

```python
dl_dx = tf.gradients(NLLCriterion, [x])[0]
```

```python
sess.run(dl_dx,feed_dict = temp_dict)
```

#### Defining our optimization prodedure

```python
learning_rate = .15
momentum = .1
```

```python
# train_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(NLLCriterion)
# train_step = tf.train.AdagradOptimizer(learning_rate).minimize(NLLCriterion)
train_step = tf.train.MomentumOptimizer(learning_rate,momentum).minimize(NLLCriterion)
```

```python
sess.run(tf.global_variables_initializer())
sess.run(tf.trainable_variables())
```

#### We can use these to evaluate accuracy of label assignment

```python
model_label = tf.argmax(softmax_y,-1)
true_label = tf.argmax(y_,-1)
```

```python
cost_tracker = pd.DataFrame([None,None,None,None],
                                  index = ['Iteration','Training Cost','Training Precision', 'Validation Cost']).T
iter_count = 0
losses = []
```

```python
training_epochs = 5
steps_per_train_epoch = data_sets.train.num_examples // batch_size
```

```python
for i in range(training_epochs):
    for j in range(steps_per_train_epoch):
        batch_dict = fill_feed_dict(data_sets.train, x, y_, batch_size)
        sess.run(train_step,feed_dict = batch_dict)
        
        iter_count += 1
        batch_loss = sess.run(NLLCriterion, feed_dict = batch_dict)
        losses.append(batch_loss)

        if (iter_count) % 500 == 0:
            
            train_cost = sess.run(NLLCriterion, feed_dict = fill_feed_dict(data_sets.train, x, y_,data_sets.train.num_examples))
            valid_cost = sess.run(NLLCriterion, feed_dict = fill_feed_dict(data_sets.validation, x, y_,data_sets.validation.num_examples))
            
            correct = tf.equal(tf.argmax(softmax_y,1), tf.argmax(y_,1))
            accuracy = tf.reduce_mean(tf.cast(correct,tf.float32))
            acc = sess.run(accuracy,feed_dict = fill_feed_dict(data_sets.train, x, y_,data_sets.train.num_examples))
            
            ct_row = pd.DataFrame([iter_count+1,train_cost,acc,valid_cost],
                                  index = ['Iteration','Training Cost','Training Accuracy', 'Validation Cost']).T
            cost_tracker = cost_tracker.append(ct_row)

            print("Iteration:", '%04d' % (iter_count),"Training Accuracy=","{:.9f}".format(acc),"Batch Loss=","{:.9f}".format(batch_loss))
            print("Training cost=", "{:.9f}".format(train_cost), "Validation cost=", "{:.9f}".format(valid_cost))
```

```python
plt.plot(cost_tracker['Iteration'],cost_tracker['Training Cost'])
plt.plot(cost_tracker['Iteration'],cost_tracker['Validation Cost'])
plt.legend()
plt.show()
```

```python
plt.plot(cost_tracker['Iteration'],cost_tracker['Training Accuracy'])
plt.legend()
plt.show()
```

```python
sess.run(accuracy, feed_dict = fill_feed_dict(data_sets.test, x, y_, data_sets.test.num_examples))
```
